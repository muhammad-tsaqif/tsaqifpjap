```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
# Initialization
n_particle = 10
n_iteration = 100
x_pos = np.zeros((n_particle, n_iteration))
y_pos = np.zeros((n_particle, n_iteration))

# Iteration
for i in range(1, n_iteration):
  for j in range(n_particle):
    rand = random.uniform(0, 1)
    # Kanan
    if rand <= 0.125:
      x_pos[j][i] = x_pos[j][i - 1] + 1
      y_pos[j][i] = y_pos[j][i - 1]
    # Kanan Bawah (serong bawah kanan)
    elif rand <= 0.250:
      x_pos[j][i] = x_pos[j][i - 1] + 1
      y_pos[j][i] = y_pos[j][i - 1] - 1
    # Bawah
    elif rand <= 0.375:
      x_pos[j][i] = x_pos[j][i - 1]
      y_pos[j][i] = y_pos[j][i - 1] - 1
    # Bwah Kiri (serong bawah kiri)
    elif rand <= 0.500:
      x_pos[j][i] = x_pos[j][i - 1] - 1
      y_pos[j][i] = y_pos[j][i - 1] - 1
    # Kiri
    elif rand <= 0.625:
      x_pos[j][i] = x_pos[j][i - 1] - 1
      y_pos[j][i] = y_pos[j][i - 1]
    # Kiri Atas (kiri serong atas)
    elif rand <= 0.750:
      x_pos[j][i] = x_pos[j][i - 1] - 1
      y_pos[j][i] = y_pos[j][i - 1] + 1
    # Atas 
    elif rand <= 0.875:
      x_pos[j][i] = x_pos[j][i - 1]
      y_pos[j][i] = y_pos[j][i - 1] + 1
    # Atas Kanan (serong atas kanan)
    else:
      x_pos[j][i] = x_pos[j][i - 1] + 1
      y_pos[j][i] = y_pos[j][i - 1] + 1


plt.figure(figsize=(10,10))

for i in range(n_particle):
  for j in range(1, n_iteration):
    plt.plot(x_pos[i], y_pos[i])

plt.title("Graph Random Walk 2D")
plt.show()
```
